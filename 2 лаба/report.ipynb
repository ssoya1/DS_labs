# Звіт про Аналіз Нерухомості


## Залежність ціни продажу від класу оподаткування


**Типи оподаткування**
 - Клас 1: Включає більшість житлової нерухомості до трьох одиниць (наприклад, одно-, дво- і трисімейні будинки, а також невеликі магазини або офіси з однією або двома прибудованими квартирами), вільні земельні ділянки, відведені під житлову забудову, і більшість кондомініумів, які не перевищують трьох поверхів.
 - Клас 2: Включає всю іншу нерухомість, яка є переважно житловою, наприклад, кооперативи та кондомініуми.
 - Клас 3: включає нерухомість з обладнанням, що належить газовим, телефонним або електричним компаніям.
 - Клас 4: Включає всі інші об'єкти нерухомості, що не увійшли до класів 1, 2 і 3, такі як офіси, фабрики, склади, гаражі тощо.

Як бачимо, в усіх районах найвищі середні ціни має нерухомість, яка підлягає 4-му класу оподаткування, що вказує на великий попит на нежитлову нерухомість. Слідом за цим, найбільш високу середню ціну має 2-й тип оподаткування, переважно будівлі типу B (перелік типів будівель: https://www.nyc.gov/assets/finance/jump/hlpbldgcode.html), що вказує на попит на двосімейну житлову нерухомість. На третьому місці знаходиться нерухомість 1-го типу оподаткування, а найменшу ціну має 3-й тип оподаткування, що говорить про те, що нерухомість цього типу є найменш привабливою.
Як можна побачити на графіках - ця тенденція є актуальної для всіх районів. 

**Висновок:** Деякі види оподаткування можуть робити інвестування в нерухомість більш або менш привабливим. Наприклад, податкові пільги або субсидії для інвесторів можуть стимулювати попит і підтримувати вищі ціни на нерухомість.

## Залежність ціни продажу від типу будівлі


Перелік типів будівель: https://www.nyc.gov/assets/finance/jump/hlpbldgcode.html

Найбільший попит, окрім Бронксу, Брукліну та Стейтен Айленду, присутній на нерухомість офісного типу, а саме офіси в будові 20 поверхів та вище для Мангеттену та офіси від 7 до 19 поверхів в районі Куінс. Подивившись на графіки та подивившись до переліку типів - можна побачити, що в районах Куінс та Мангеттен більшим попитом користуються офісні будови, а тому ці райони є переважно комерційними, до чого можна додати що Мангеттен є більш урбанізованим через розмір будівель. В той самий час район Бронкс є переважно промисловим, бо саме фабрики та транспортна нерухомість має там найбільший попит. У Брукліні найбільшим попитом користується освітня нерухомість, а в районі Стейт Айленд - торгівельна.

**Висновок:** Різні типи будівель мають різні функціональні цілі, такі як житлова нерухомість, комерційна нерухомість (наприклад, офіси, магазини, склади), промислові об'єкти тощо. Ціни на ці типи будівель можуть сильно відрізнятися через різні потреби і попит на різні види нерухомості.

## Кількість будинків за роками будівництва


Як можна побачити за графіками, найбільша кількість будинків буда побудована в першій половині ХХ-го сторічча, окрім району Стейт Айленд, де пік будивництва припав на його кінець.

**Висновок:** Висока кількість нових будівель може свідчити про живий або зростаючий ринок нерухомості. Це може вказувати на активність у будівництві, зростання попиту на житло або комерційні об'єкти, що може бути пов'язано зі сприятливою економічною ситуацією в регіоні. Активність у будівництві нових будинків може слугувати індикатором економічної активності в даному регіоні. Це може впливати на різні сфери економіки, включаючи будівельну, видобувну промисловість, а також споживання будівельних матеріалів.

## Залежність ціни продажу від площі земельної ділянки (квадратні фути)


**Висновок:** Ціна земельної ділянки може залежати від її місця розташування. Ділянки в центрі міста або з хорошою транспортною доступністю можуть мати вищу ціну порівняно з ділянками на околицях або у віддалених районах, і це помітно на графікаї районів Мангеттену та Куінс, де ціна є найвищою. Наявність уже наявної інфраструктури на ділянці або її близькість до зручностей, таких як школи, лікарні, магазини, також може впливати на ціну земельної ділянки.

## Залежність ціни продажу від площі будівлі (квадратні фути)


**Висновок:** Часто ціна нерухомості безпосередньо корелює з її площею. Це дає змогу покупцям та інвесторам порівнювати різні об'єкти на ринку за вартістю за одиницю площі та ухвалювати більш поінформовані рішення.Зміни в ціні за одиницю площі будівлі можуть вказувати на тенденції на ринку нерухомості. Наприклад, зниження ціни за квадратний метр може вказувати на зміну попиту на певний вид нерухомості або на надлишок пропозиції. 

## Залежність середньої ціни від року будівництва


**Висновок:** Залежність ціни від року будівництва може показати різницю в цінах між старими та новими об'єктами. Зазвичай нові будівлі мають вищу ціну через більш сучасні зручності, технології та кращий стан. У різних регіонах ціни на нерухомість можуть змінюватися залежно від року будівництва. Наприклад, у деяких місцях старі будівлі можуть мати високу ціну через історичну цінність або унікальність. Як можемо побачити, в Бронксі та у Брукліні найдорожчими будівлями на ринку є будинки побудовані на початку ХХ-го сторічча, в той час як на Мангеттені та у районі Куінс - в другі половині. Тим часом ціни на Стейт Айленді розподілені більш рівнмірно.

## Залежність ціни від сусідства






**Висновок:** Нерухомість, що розташована поруч із ключовими об'єктами, такими як школи високого рейтингу, парки, торговельні центри, громадський транспорт або центри зайнятості, може мати вищу вартість через зручність і доступність таких об'єктів. Наявність добре розвиненої інфраструктури в околицях, такої як дороги, системи громадського транспорту, магазини, ресторани та інші зручності, може підвищувати цінність нерухомості в цьому районі. Деякі райони або сусідства можуть бути більш престижними та ексклюзивними, що може призводити до вищих цін на нерухомість через їхню репутацію та статус.

## Кореляція до ціни


**Висновок:**Кореляція ціни нерухомості з певною характеристикою може надати інформацію про ступінь взаємозв'язку між цією характеристикою і ціною. Зазвичай кореляція дає змогу зрозуміти, наскільки сильно або слабо дві змінні взаємопов'язані. У разі нерухомості, характеристики можуть варіюватися від площі приміщення або земельної ділянки до року побудови, розташування в конкретному районі та інших факторів. За графіком можна зробити висновок, що на ціну нерухомості у Бронксі найбільше впливають загальна кількість нерухомості та наявність житла, в той час як на Мангеттені, у Куінсі та на Стейт Айленді найбільш впливає площа будівлі, а в Брукліні - наявність комерційної нерухомості.

## Коваріація до ціни


**Висновок:** Коваріація ціни нерухомості та конкретної характеристики в контексті ринку нерухомості може надати деяку інформацію про напрямок зв'язку між цією характеристикою та ціною. Коваріація може вказувати на те, які характеристики або фактори можуть впливати на ціну нерухомості. Наприклад, позитивна коваріація між ціною та площею приміщення може означати, що великі приміщення часто продаються за вищими цінами. Як можемо побачити за графіком, найбільший коваріативний вплив у більшості районів, окрім Бронксу та Стейт Айленду надає площа самої будівлі, в той час як в зазначених районах такий впил більше надає площа землі нерухомості.

## Дисперсія
  

**Висновок:** Дисперсія ціни нерухомості являє собою міру мінливості або розкиду цін на ринку нерухомості. Це статистичне значення дає змогу оцінити, наскільки сильно різняться ціни нерухомості щодо їхнього середнього значення.Аналіз дисперсії може допомогти в класифікації ринку нерухомості. Ринок із низькою дисперсією може бути більш стійким і передбачуваним, тоді як висока дисперсія може вказувати на більш різноманітний і мінливий ринок. За графіком можемо сказати, що найпередбачуванішим є район Куінс, а найнебередбачуваним - Стейт Айленд.
