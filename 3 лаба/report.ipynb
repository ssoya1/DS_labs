# Аналіз взаємозалежності факторів

## Залежність числових даних між собою
Для аналізу взаємозв'язків було використано діаграми розсіювання (scatter plots) для параметрів, занесених до списку features, таких як довжина автомобіля, ширина, висота, розмір двигуна, витрата пального в місті та на шосе, а також інші технічні характеристики, наявні в наборі даних.

Проведений аналіз виявив кілька важливих взаємозв'язків між різними параметрами старих автомобілів. Наприклад, виявлено позитивну кореляцію між розміром двигуна (engine-size) і вагою автомобіля (curb-weight), що може вказувати на те, що автомобілі з більшими двигунами мають тенденцію до більшої ваги.

**Висновок:** Аналіз дав змогу виявити кілька важливих взаємозв'язків між різними параметрами технічних характеристик старих автомобілів. Розуміння цих взаємозв'язків може бути корисним для визначення структури даних і може допомогти в подальшому ухваленні рішень про ціноутворення або визначенні основних чинників, що впливають на технічні характеристики автомобілів.










## Аналізу взаємозв'язків між чисельними ознаками
Для аналізу взаємозв'язків між чисельними ознаками було використано кореляційну матрицю, яка відображає ступінь лінійного зв'язку між парами ознак. Значення кореляції коливаються від -1 до 1: ближче до 1 вказує на сильний позитивний зв'язок, ближче до -1 - на сильний негативний зв'язок, а значення близьке до 0 означає відсутність лінійної залежності.

Проведений аналіз кореляційної матриці дав змогу виявити кілька важливих взаємозв'язків між чисельними параметрами автомобілів. Наприклад, виявлено позитивну кореляцію між параметрами витрати пального в місті та на шосе. Це вказує на те, що автомобілі з вищою витратою пального в місті також схильні до мати вищу витрату пального на шосе, і навпаки.

Також спостерігається позитивна кореляція між розміром двигуна (engine-size) і вагою автомобіля (curb-weight), що може вказувати на те, що автомобілі з більшим розміром двигуна схильні до мати вищу вагу.

**Висновок:**Аналіз кореляційної матриці допоміг виявити важливі взаємозв'язки між чисельними параметрами старих автомобілів. Це розуміння може бути корисним для подальшого аналізу даних і ухвалення рішень, наприклад, для ідентифікації ключових чинників, що впливають на технічні характеристики автомобілів.



## Відношення ціни до числових властивостей
Було побудовано графіки розсіювання (scatter plots) для чисельних ознак щодо ціни автомобіля, а також boxplot'и для візуального аналізу розподілу кожної ознаки.

Графіки розсіювання показали різноманітні залежності між чисельними параметрами та ціною автомобілів. Як можемо спостерігати на графіку, кореляція між ціною автомобілів та їх параметрами - відсутня, що говорить про те, що зазначені параметри можуть не мати залежності з ціною автомобіля в цьому наборі даних.
Boxplot'и дозволили побачити основні характеристики розподілу чисельних ознак. Більшість параметрів мають різний розподіл значень.

**Висновок:**Аналіз графіків розсіювання і boxplot'ів дав змогу виявити різні взаємозв'язки між чисельними ознаками і ціною автомобілів. Ці результати можуть бути корисними для розуміння впливу технічних характеристик на ціну вживаних автомобілів.




## Перевірка на викиди
Для чисельних ознак автомобілів було застосовано метод Winsorize з параметрами limits=[0.05, 0.05], який обрізає крайні значення, встановлюючи їх рівними певним квантилям даних. Далі було проведено аналіз отриманих даних через побудову boxplot'ів для візуалізації змін у розподілі значень після застосування методу Winsorize.

Після застосування методу Winsorize для опрацювання крайніх значень чисельних параметрів автомобілів, помітні зміни в розподілі даних. Boxplot'и після застосування Winsorize показують, що викиди в даних було пом'якшено або видалено, що призвело до більш врівноваженого розподілу значень за різними ознаками.

**Висновок:**Застосування методу Winsorize дало змогу зменшити вплив викидів на чисельні ознаки автомобілів і зробило розподіл даних більш стабільним і стійким до екстремальних значень.



## Використання методів з прикладу
Було застосовано такі методи відбору ознак з прикладу:

- Використання методу VarianceThreshold для видалення ознак із низькою дисперсією.
- Вибір ознак із найбільшою кореляцією з цільовою змінною (F-критерій).
- Застосування методу Lasso для вибору найважливіших ознак.
- Використання моделі RandomForestRegressor для оцінки важливості ознак.

Графік демонструє вплив однієї ознаки на іншу за різних значень. Червоні відтінки вказують на позитивну кореляцію, сині - на негативну, а прозорість - на відсутність впливу. Наприклад, якщо вага curb-weight висока, то найімовірніше і розмір двигуна (engine-size) також буде більшим. Це ж спостерігається і для інших ознак з аналогічними значеннями. Також можна відзначити пари ознак, взаємний вплив яких є негативним, наприклад, engine-location, перетворений із текстового в числовий формат, що містить значення front і rear. Якщо значення першої ознаки, то значення compression-ratio більше, і навпаки.

**Висновок:**Результати ранжування дають змогу виокремити найбільш значущі ознаки для побудови моделі прогнозування ціни автомобіля. Виходячи з цих висновків, рекомендується використовувати виділені важливі ознаки для підвищення точності моделей прогнозування ціни старих автомобілів. Також варто зазначити, що різні методи відбору ознак можуть давати різні результати залежно від конкретних особливостей даних і цілей моделювання.
